```python
import pandas as pd
import json
```

```python
encountersdf = pd.read_csv('../data/encounter_data_new.csv')
```

```python
encountersdf.columns = ['ID', 'Age', 'Gender', 'Race', 'City', 'State', 'lat', 'lng',  'Cause_of_Death', 'Mental_Illness', 'Year']
encountersdf["City/State"] = encountersdf["City"] + "_" + encountersdf["State"]
encountersdf["City/State"] = encountersdf["City/State"].str.replace(' ', '')
encountersdf["City/State"] = encountersdf["City/State"].str.replace('-', '')
encountersdf["City/State"] = encountersdf["City/State"].str.replace("'", "")
encountersdf["City/State"] = encountersdf["City/State"].str.replace(".", "")
encountersdf["City/State"] = encountersdf["City/State"].str.replace("(", "")
encountersdf["City/State"] = encountersdf["City/State"].str.replace(")", "")
encountersdf
```

```python
encounter_by_city = encountersdf.groupby("City/State").agg({"lat": ["first"], "lng": ["first"], "Cause_of_Death": ["count"]})
encounter_by_city = encounter_by_city.reset_index()
encounter_by_city.columns = ["City/State","lat", "lng", "count"]
encounter_by_city["count"] = encounter_by_city["count"]*500
```

```python
markerdata = []
for i in range(len(encounter_by_city)) :
   marker = f"var {encounter_by_city['City/State'][i]} = L.circle([{encounter_by_city['lat'][i]}, {encounter_by_city['lng'][i]}],< color: 'red', fillColor: '#03', fillOpacity: 0.5, radius:{encounter_by_city['count'][i]}>).addTo(myMap);"
   markerdata.append(marker)
```

```python
markerdata
```

```python
with open('encounter.json', 'w') as fp:
    json.dump(markerdata, fp)
```
