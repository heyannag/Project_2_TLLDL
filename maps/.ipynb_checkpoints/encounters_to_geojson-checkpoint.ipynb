```python
import pandas as pd
import json
```

```python
encountersdf = pd.read_csv('../data/encounter_data_new.csv')
```

```python
encountersdf.columns = ['ID', 'Age', 'Gender', 'Race', 'City', 'State', 'lat', 'lng',  'Cause_of_Death', 'Mental_Illness', 'Year']
encountersdf.head()
```

```python
markerdata = []
for i in range(len(encountersdf)) :
   marker = f"{encountersdf['State'][i]} = L.marker([{encountersdf['lat'][i]}, {encountersdf['lng'][i]}]).bindPopup('COD':{encountersdf['Cause_of_Death'][i]})"
   markerdata.append(marker)
```

```python
markerdata
```

```python
with open('encounter.json', 'w') as fp:
    json.dump(markerdata, fp)
```
